```python
# 1️⃣ Import libraries
import pandas as pd

# 2️⃣ Read the CSV (replace with your uploaded file name if different)
df = pd.read_csv('sales_data.csv')

# 3️⃣ Quick check
print("First 5 rows:")
display(df.head())

print("\nColumn info:")
print(df.info())

# 4️⃣ Clean numeric columns
# Only 'Sales' exists as numeric
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# 5️⃣ Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# 6️⃣ Basic SQL-like analysis

# Total sales
total_sales = df['Sales'].sum()
print("\nTotal Sales:", total_sales)

# Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Region:")
print(sales_by_region)

# Sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Category:")
print(sales_by_category)

# Top 10 Products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products:")
print(top_products)

# Monthly sales trend
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
print("\nMonthly Sales Trend:")
print(monthly_sales)

# 7️⃣ Export cleaned CSV (optional, for Power BI)
df.to_csv('sales_data_cleaned.csv', index=False)
print("\nCleaned CSV exported as 'sales_data_cleaned.csv'")
```
